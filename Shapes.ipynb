```python
import matplotlib.pyplot as plt
import numpy as np

Cx = float(input("Enter the X coordinate of the centre of the circle"))
Cy = float(input("Enter the Y coordinate of the centre of the circle"))
r = float(input("Enter the Radius of the circle"))

theta = np.linspace(0, 2 * np.pi, 100)
x = Cx + r * np.cos(theta)
y = Cy + r* np.sin(theta)

plt.figure()
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Circle')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

h = float(input("Enter the X coordinate of the centre of the circle"))
k = float(input("Enter the Y coordinate of the centre of the circle"))
r = float(input("Enter the Radius of the circle"))

x = np.linspace(h - r, h + r, 400)

# Calculate corresponding y-coordinates using the circle formula
y_positive = k + np.sqrt(r**2 - (x - h)**2)
y_negative = k - np.sqrt(r**2 - (x - h)**2)

plt.figure(figsize=(5, 5))
plt.plot(x, y_positive)
plt.plot(x, y_negative)
plt.gca().set_aspect('equal', adjustable='box')  # Set aspect ratio to make it a perfect circle
plt.title('Circle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the parameters of the ellipse
Cx = float(input("Enter the X coordinate of the centre of the Ellipse"))
Cy = float(input("Enter the Y coordinate of the centre of the Ellipse "))
SMaj = float(input("Enter the Semi Major Axis of the Ellipse"))
SMin = float(input("Enter the Semi Minor Axis of the Ellipse"))
angle = 30  # Angle of rotation in degrees

# Generate points for the ellipse
theta = np.linspace(0, 2 * np.pi, 100)
x = Cx * SMaj *np.cos(np.radians(angle)) * np.cos(theta) - SMin * np.sin(np.radians(angle)) * np.sin(theta)
y = Cy * SMaj * np.sin(np.radians(angle)) * np.cos(theta) + SMin* np.cos(np.radians(angle)) * np.sin(theta)

# Create a plot
plt.figure()
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Ellipse')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the hyperbola parameters
a = float(input("Enter the Horizontal axis semi-length"))
b = float(input("Enter the Vertical axis semi-length"))
h = float(input("Enter the x-coordinate of the center"))
k = float(input("Enter the y-coordinate of the center"))

x = np.linspace(h - 3 * a, h + 3 * a, 400)

y_positive = k + (b / a) * np.sqrt(a**2 + (x - h)**2)
y_negative = k - (b / a) * np.sqrt(a**2 + (x - h)**2)

plt.figure(figsize=(8, 6))
plt.plot(x, y_positive, label='Upper Branch', color='blue')
plt.plot(x, y_negative, label='Lower Branch', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hyperbola')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the parameters of the parabola
a = int(input("Coefficient that determines the width of the parabola"))  # Coefficient that determines the "width" of the parabola
b = int(input("Coefficient that determines the horizontal shift"))  # Coefficient that determines the horizontal shift
c = int(input("Coefficient that determines the vertical shift"))  # Coefficient that determines the vertical shift

# Create an array of x-values
x = np.linspace(-10, 10, 400)

# Calculate the corresponding y-values using the quadratic equation
y = a * x**2 + b * x + c

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the parabola
ax.plot(x, y, label='Parabola', color='blue')

# Set axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Add a legend
ax.legend()

# Show the plot
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
